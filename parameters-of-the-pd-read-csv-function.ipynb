Understanding the various parameters of the `pd.read_csv()` function is essential for harnessing its full potential and effectively handling different data scenarios. In this notebook, we will explore the key parameters of the `pd.read_csv()` function and discuss how they can be utilized to customize data loading, handle different file formats, and address common data challenges.

# <a name="1"></a><h1 style="font-size: 36px; font-family: 'Roboto Mono', monospace; font-weight: normal; background-color: #edf2fb; color:#ffa62b; text-align: center; border-radius: 50px 50px;"><b>1. Header parameter</b></h1>

- Specifies the row(s) to use as the column names. 
- By default, the first row of the file is used as the header. 
- Using this parameter, you can specify a different row as the header 

```python
df = pd.read_csv('/kaggle/input/test-csv/test.csv')
df
```

here the row 1 is the names of the column, so to make it as the column name, we can pass in `header=1`

```python
pd.read_csv('/kaggle/input/test-csv/test.csv', header=1)
```

# <a name="1"></a><h1 style="font-size: 36px; font-family: 'Roboto Mono', monospace; font-weight: normal; background-color: #edf2fb; color:#ffa62b; text-align: center; border-radius: 50px 50px;"><b>2. usecols parameter</b></h1>
- Selects specific columns to read from the file. 
- Instead of reading all columns, you can provide a list of column names or column indices to include only those columns in the resulting DataFrame.

```python
df1 = pd.read_csv('/kaggle/input/world-air-quality-index-by-city-and-coordinates/AQI and Lat Long of Countries.csv')
df1.head()
```

```python
pd.read_csv('/kaggle/input/world-air-quality-index-by-city-and-coordinates/AQI and Lat Long of Countries.csv',
            usecols=['Country', 'City', 'AQI Value', 'lat', 'lng'])
```

# <a name="1"></a><h1 style="font-size: 36px; font-family: 'Roboto Mono', monospace; font-weight: normal; background-color: #edf2fb; color:#ffa62b; text-align: center; border-radius: 50px 50px;"><b>3. squeeze parameter</b></h1>
- used to import just one column as **a series** out of all columns

```python
series_squeeze = pd.read_csv('/kaggle/input/world-air-quality-index-by-city-and-coordinates/AQI and Lat Long of Countries.csv', 
            usecols=['Country'], squeeze=True)

series_squeeze
```

```python
type(series_squeeze)
```

# <a name="1"></a><h1 style="font-size: 36px; font-family: 'Roboto Mono', monospace; font-weight: normal; background-color: #edf2fb; color:#ffa62b; text-align: center; border-radius: 50px 50px;"><b>4. skiprow parameter</b></h1>
- used to skip certain rows
- the row number to be skipped has to be passed as a list
- can also use lambda functions and other complex logic to skip the rows accordingly

```python
pd.read_csv('/kaggle/input/world-air-quality-index-by-city-and-coordinates/AQI and Lat Long of Countries.csv',
            skiprows=[1,5,10])
```

```python
pd.read_csv('/kaggle/input/world-air-quality-index-by-city-and-coordinates/AQI and Lat Long of Countries.csv',
            skiprows=(lambda x: x%2==0))

#skipped all the even indexed rows; you can use the 'names' parameter and specify the header 
```

# <a name="1"></a><h1 style="font-size: 36px; font-family: 'Roboto Mono', monospace; font-weight: normal; background-color: #edf2fb; color:#ffa62b; text-align: center; border-radius: 50px 50px;"><b>5. nrows parameter</b></h1>
- used to just import `n` rows
- can be useful when working with large datasets and you want to load only a subset of the data for initial exploration.

```python
pd.read_csv('/kaggle/input/world-air-quality-index-by-city-and-coordinates/AQI and Lat Long of Countries.csv', nrows=100)
```

# <a name="1"></a><h1 style="font-size: 36px; font-family: 'Roboto Mono', monospace; font-weight: normal; background-color: #edf2fb; color:#ffa62b; text-align: center; border-radius: 50px 50px;"><b>6. encoding parameter & error_bad_lines parameter</b></h1>

### encoding parameter
- sometimes it so happens that some datasets have a different encoding than utf-8
- in this cases, we use the encoding parameter to pass the encoding format in which the dataset is encoded in, so that pandas can parse it properly
- if you do not know the encoding format, you can open the dataset on any text editor and change its encoding format to utf-8

### error_bad_lines parameter
- sometimes it so happens that there mighe be more values in the rows than the total columns in the dataset
- in such cases, pandas throws an error
- to avoid such rows, we can set the `error_bad_lines parameter` to `False`

```python
pd.read_csv('/kaggle/input/character-encoding-examples/yan_BIG-5.txt', encoding='Big5', error_bad_lines=False)
```

# <a name="1"></a><h1 style="font-size: 36px; font-family: 'Roboto Mono', monospace; font-weight: normal; background-color: #edf2fb; color:#ffa62b; text-align: center; border-radius: 50px 50px;"><b>7. converters parameter</b></h1>
- Applies custom functions to specific columns during parsing
- sometimes we might want to do some convertions to the data right before we import it, in such cases, `convertors` parameter can be used
<br>

**Example**
- suppose we want to convert 'Russian Federation' to 'Russia' and 'Brazil' to 'BR', we can use the converters parameter while reading the file and perform the operation

```python
def rename(name):
    if name == "Russian Federation":
        return "Russia"
    elif name == 'Brazil':
        return 'BR'
    else:
        return name
```

```python
pd.read_csv('/kaggle/input/world-air-quality-index-by-city-and-coordinates/AQI and Lat Long of Countries.csv', 
            converters={'Country':rename})

# as you can see Russia and Brazil have been renamed
```

# <a name="1"></a><h1 style="font-size: 36px; font-family: 'Roboto Mono', monospace; font-weight: normal; background-color: #edf2fb; color:#ffa62b; text-align: center; border-radius: 50px 50px;"><b>8. chunksize parameter</b></h1>
- Enables reading the file in smaller chunks or chunks of specified size
- Instead of loading the entire file into memory at once, this parameter allows you to process the data in manageable portions.
- By setting the chunksize, you can iterate over the file and perform operations on each chunk individually, which is particularly useful for working with large datasets that cannot fit into memory entirely.

```python
dfs = pd.read_csv('/kaggle/input/world-air-quality-index-by-city-and-coordinates/AQI and Lat Long of Countries.csv',chunksize=5000)
```

```python
for chunks in dfs:
    print(chunks.shape)
```

Now you can apply all the normal pandas operations inside the `for` loop, and it would get applied on each chunk
